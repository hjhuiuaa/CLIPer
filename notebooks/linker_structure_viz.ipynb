# CLIPer linker 对比可视化

- linker 残基统一高亮为香芋紫（`#BFA6E8`）
- HTML 已包含三视图：Predicted / True / Pred-vs-True 比对
- 本 notebook 负责快速浏览批量输出结果

```python
from pathlib import Path
import csv
from IPython.display import IFrame, display

OUT_DIR = Path("../artifacts/structure_viz").resolve()
MANIFEST = OUT_DIR / "manifest.tsv"
HTML_DIR = OUT_DIR / "html"

rows = []
with MANIFEST.open("r", encoding="utf-8", newline="") as f:
    rows = list(csv.DictReader(f, delimiter="\t"))

print(f"manifest: {MANIFEST}")
print(f"records: {len(rows)}")
rows[:5]
```

```python
def show_compare(protein_id: str, width: int = 1400, height: int = 900):
    html_path = HTML_DIR / f"{protein_id}.html"
    if not html_path.exists():
        raise FileNotFoundError(f"HTML not found: {html_path}")
    display(IFrame(src=html_path.as_uri(), width=width, height=height))

# 示例
# show_compare("DP04156")
```
